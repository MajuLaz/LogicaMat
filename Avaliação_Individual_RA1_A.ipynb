# Questão 1: 
A população de uma cidade está crescendo exponencialmente. No ano de 2020, a população era de 50.000 habitantes, e a taxa de crescimento anual é de 3%. Qual será a população em 2030? Use a fórmula:

$$
P(t) = P_0 \cdot (1 + r)^t
$$

onde $P_0$ é a população inicial, $r$ é a taxa de crescimento e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando o crescimento populacional entre os anos de 2020 e 2040, ano a ano.


```python
import sympy as sp
import matplotlib.pyplot as plt

t = sp.symbols('t')  
Po = 50000  
r = 0.03  
P = Po * (1 + r)**t
P.subs(t, 10)
P_func = sp.lambdify(t, P)
anos = range(2020, 2041)
populacoes = [P_func(i - 2020) for i in anos]
plt.plot(anos, populacoes)
plt.xlabel('Ano')
plt.ylabel('População')
plt.title('Crescimento populacional exponencial')
plt.show()
```

# Questão 2
Usando, a seu critério as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $2^x + 3^x = 25$

2. $\log_2(x) - \log_2(3) = 1$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 25 - 2*x - 3*x

x = np.linspace(-5, 5, 100)

plt.plot(x, f(x))
plt.xlabel('$2^x+3^x=25$')
plt.ylabel('y')
plt.title('Função decrescente')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 10, 1000)  
y = np.log2(x) - np.log2(3) - 1  

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('$log_2(x) - log_2(3) - 1$')
plt.title('Função logarítmica decrescente')
plt.grid()
plt.show()
```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 2} \frac{x^3 - 8}{x - 2}$

2. $\lim_{x \to \infty} \frac{3x^2 + 2x - 1}{x^2 - 4x + 7}$



1. f(x) = (x^3 - 8)/(x - 2)
   f(x) = ((x - 2)(x^2 + 2x + 4))/(x - 2)
   f(x) = x^2 + 2x + 4
   lim x -> 2 (x^3 - 8)/(x - 2) = lim x -> 2 (x^2 + 2x + 4) = 12


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 1000)  
y = (x**3 - 8)/(x - 2)  

plt.plot(x, y)
plt.axvline(x=2, color='red', linestyle='--')  
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.grid()
plt.show()
```

2. lim x -> ∞ (3x ^ 2 + 2x - 1)/(x ^ 2 - 4x + 7)
   lim x -> ∞ (3x^2 + 2x - 1)/(x^2 - 4x + 7) = lim x -> ∞ (6x + 2)/(2x - 4)
   lim x -> ∞ (3x^2 + 2x - 1)/(x^2 - 4x + 7) = lim x -> ∞ (6x + 2)/(2x - 4) = lim x -> ∞ 3

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)  
y = (3*x*2 + 2*x - 1)/(x*2 - 4*x + 7)  


plt.plot(x, y)
plt.axhline(y=3, color='red', linestyle='--')  
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.grid()
plt.show()
```
